# If – Elif – Else Statements

## If Elif Else Flowchart

```mermaid
flowchart TD
    A[Start] --> B{If condition}
    B -- True --> C[Body of if]
    B -- False --> D{Elif condition}
    D -- True --> E[Body of elif]
    D -- False --> F[Body of else]
    C --> G[End]
    E --> G
    F --> G


````python
if 20 < 30:
    print('If worked')
````

````python
if 20 < 30:
    print('If worked')
else: 
    print('If do not work')
````

````python
if 20 < 10:
    print('If worked')
else: 
    print('If do not work')
````

````python
if 20 < 10:
    print('If worked')
elif 25 < 21:
    print('elif 1 worked')
elif 25 < 22:
    print('elif 2 worked')
elif 25 < 30    :
    print('elif 3 worked')
elif 25 < 40:
    print('elif 4 worked')
    
else: 
    print('If do not work')
````

````python
if (20 < 10) or (20 < 30):
    print('If worked')
elif 25 < 21:
    print('elif 1 worked')
elif 25 < 22:
    print('elif 2 worked')
elif 25 < 30    :
    print('elif 3 worked')
elif 25 < 40:
    print('elif 4 worked')
    
else: 
    print('If do not work')
````

````python
print('It worked') if 20 > 30 else print('It did not work...')
````

````python
print('It worked') if  20 > 10 else print('It dod not work...')
````

````python
print('It worked') if 20 < 10 else( 
    print('It elif worked...') if 20 < 30
    else print('It not work...')
    )
````

````python
print("It worked") if 20 < 10 else (
    print("It elif worked") if 20 < 30 else print("It did not work")
)
````

````python
if (20 < 10) or (20 < 30):
    print('If worked')
    if 10 > 5:
        print('This nested if statement worked')
elif 25 < 21:
    print('elif 1 worked')
elif 25 < 22:
    print('elif 2 worked')
elif 25 < 30    :
    print('elif 3 worked')
elif 25 < 40:
    print('elif 4 worked')
    
else: 
    print('If do not work')
````
