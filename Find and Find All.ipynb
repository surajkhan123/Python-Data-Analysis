# Find and Find All

## Web Scraping 

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.w3schools.com/html/html_examples.asp'
```

```python
requests.get(url)
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('h1')
```

```python
soup.find_all('h1')
```

```python
soup.find('div')
```

```python
soup.find_all('div')
```

```python
soup.find('p')
```

```python
soup.find('p', class_ = 'tnb-services-headlines')
```

```python
soup.find_all('p', class_ = 'tnb-services-headlines')
```

```python
soup.find('p', class_ = 'tnb-services-headlines').text
```

```python
soup.find_all('p', class_ = 'tnb-services-headlines').text
```

```python
soup.find('p', class_ = 'tnb-services-headlines').text
```

```python
soup.find('p', class_ = 'tnb-services-headlines').text.strip()
```
