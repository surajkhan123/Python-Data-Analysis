# For Loop

## For Loop Flowchart

```mermaid
flowchart TD
    A([Start]) --> B[Entering for loop<br/>for each item in sequence]
    B --> C{Last element reached?}
    C -- No --> D[Body of for loop]
    D --> C
    C -- Yes --> E[Exiting loop]
    E --> F[Statement after the for loop]

````python
integers = [1,2,3,4,5]
````

````python
for number in integers:
    print(number)
````

````python
for suraj in integers:
    print(suraj)
````

````python
for number in integers:
    print('Suraj')
````

````python
for Suraj in integers:
    print(Suraj + Suraj)
````

````python
ice_cream_dict = {'name': 'Suraj Khan', 'weekly intake': 5, 'favourite ice creams':['MCC','Chocolate']}
````

````python
for ice in ice_cream_dict.values():
    print(ice)
````

````python
for ice in ice_cream_dict.items():
    print(ice)
````

````python
for key, value in ice_cream_dict.items():
    print(key,'->',value)
    
````

# Nested For Loops

````python
flavors = ['Vanilla','Chocolate','Cookie Dough']
toppings = ['Hot Fudge','Oreos','Marshmallows']
````

````python
for one in flavors:
    for two in toppings:
        print(one, 'topped with', two)
````
