## Syntax of Function

```python
def first_func():
    print('We did it!')
```

```python
first_func()
```

```python
def second_func():
    print(20)
```

```python
second_func()
```

## Arguments of Functin

```python
def num(num1):
    print(num1**3)
```

```python
num(5)
```

```python
def number_squared(number):
    print(number**2)
```

```python
number_squared(3)
```

```python
def number_square_root(number):
    print(int(number ** 0.5))
```

```python
number_square_root(16)
```

## Multiple Arguments

```python
def number_squared_cust(number,power):
    print(number**power)
```

```python
number_squared_cust(5,3)
```

```python
def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(5,6)
```

## Arbitrary Arguments

```python
args_tuple = (5,6,3,2,8)

def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(*args_tuple)
```

## Keyword Arguments

```python
def number_squared_cust(number,power):
    print(number**power)
```

```python
number_squared_cust(power = 5, number = 3)
```

## Arbitrary Keyword Arguments

```python
def number_kwarg(**number):
    print('My number is: ' + number['integer'])
```

```python
number_kwarg(integer = '2025')
```

```python
def number_kwarg(**number):
    print('My Number is: ' + number['integer'])
```

```python
number_kwarg(integer = '250')
```

```python
def number_kwarg(**number):
    print('My Number is: ' + number['integer'] + ' and ' + number['integer2'])
```

```python
number_kwarg(integer = '500', integer2 = '645')
```
