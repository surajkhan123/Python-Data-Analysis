```python
daily_points_log = {1,2,31,2,3,4,1,2,5,6,3,2}
daily_points_log
```

```python
wifes_daily_points_log = {1,3,5,7,3,24,5,7,3,2,0}
wifes_daily_points_log
```

```python
# Showing all unique elements from both sets
daily_points_log | wifes_daily_points_log
```

```python
# returns only the common elements present in both sets.
daily_points_log & wifes_daily_points_log
```

```python
# Elements that are in daily_points_log BUT NOT in wifes_daily_points_log
print(daily_points_log - wifes_daily_points_log)
```

```python
# Elements that are in wifes_daily_points_log BUT NOT in daily_points_log
print(wifes_daily_points_log - daily_points_log)
```

```python
# Elements that are in either set, but NOT in both
print(wifes_daily_points_log ^ daily_points_log)
```

```python
print(daily_points_log ^ wifes_daily_points_log)
```
