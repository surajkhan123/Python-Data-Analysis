# Comparison Operators

| Operator | Name |
|---------|------|
| == | Equal |
| != | Not equal |
| > | Greater than |
| < | Less than |
| >= | Greater than or equal to |
| <= | Less than or equal to |


```python
10 == 10
```

```python
10 != 10
```

```python
'Vanilla' == 'Vanilla'
```

```python
'Vanilla' == 'Chocolate'
```

```python
'Vanilla' != 'Chocolate'
```

```python
x = 'Vanilla'
y = 'Chocolate'
x == y
```

```python
x = 'Vanilla'
y = 'Vanilla'
x == y
```

```python
x = 'Vanilla'
y = 'Chocolate'
x != y
```

```python
10 <50
```

```python
10 >= 10
```

```python
50 >= 10
```

# Logical Operators

    or                 Returns True if one of the statements is true
    not                Reverse the result, returns False if the result is true

# Membership Operators

| Operator | Description |
|---------|------------|
| in | Returns True if a sequence with the specified value is present in the object |
| not in | Returns True if a sequence with the specified value is not present in the object |


```python
ice_cream = 'I love Chocolate ice cream'
'love' in ice_cream
```

```python
ice_cream = 'I love Chocolate ice cream'
'love' in 'I love Chocolate ice cream'
```

```python
scoops = [1,2,3,4,5]

4 in scoops
```

```python
scoops = [1,2,3,4,5]

6 not in scoops
```

```python
scoops = [1,2,3,4,5]
7 in scoops
```

```python
scoops = [1,2,3,4,5]
7 not in scoops
```

```python
scoops = [1,2,3,4,5]
wanted_scoops = 8
8 in scoops
```

```python
scoops = [1,2,3,4,5]
wanted_scoops = 8
wanted_scoops in scoops
```
