# While Loop
## While Loop FLowchart

```mermaid
flowchart TD
    A([Start]) --> B[Enter the while loop]
    B --> C{Test condition}
    C -- True --> D[While loop body]
    D --> C
    C -- False --> E[Exit the while loop]
    E --> F([End])

````python
number = 0
while number < 5:
    print(number)
    number = number +1
````

````python
number = 0

while number < 5:
    print(number)
    if number == 3:
        continue
    number = number + 1
else:
    print('No longer <5')
    
````
