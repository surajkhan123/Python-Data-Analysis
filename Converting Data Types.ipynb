# Converting Data Types

```python
num_int = 7
type(num_int)
```

```python
# without quotes('') -> Number
# with quotes('') -> Text

num_str = '7'
type(num_str)
```

```python
# Convert String -> Integer
num_str_convt = int(num_str)
type(num_str_convt)
```

```python
num_sum = num_int + num_str_convt

print(num_sum)
```
